```python
#Activaton function : It is decision maker 
#AF is the very imp topics in DL .
#import basic library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Take one input (Independent variable)
#create dummy input
X=np.arange(-100,100) #start =-100 stop=100-1=99 step=+1
print(X)
```

```python
#Linear Activation : 
#Linear means straight line 
#Equation of Straigh line Y=mX+c
#m slope and c intercept 
#suppose 
m=4
c=5 
#find output Y when given input X
Y=m*X+c   #Y=4*-100+5=    Y=4*-99+5
print(Y)
```

```python
#crate a dictionary 
dict={'X':X,'Y':Y}
#convert dictionary into DataFrame 
df=pd.DataFrame(dict)
df
```

```python
sns.lineplot(data=df,x=X,y=Y)
plt.title("linear Activation function")
plt.show()
```

```python
#Sigmoid function
#formula 
Y=1/(1+np.exp(-X))  #0-1
Y=[round(i,2) for i in Y]
print(Y)
```

```python
#crate a dictionary 
dict={'X':X,'Y':Y}
#convert dictionary into DataFrame 
df=pd.DataFrame(dict)
#line plot
sns.lineplot(data=df,x=X,y=Y)
plt.title("Sigmoid Activation function")
plt.show()
```

```python
#tanh AF : -1 to +1  and Threshold value =0
#sigmoid :  0-1  and threshold value =0.5
#tanh function
#formula 
Y=(np.exp(X)-np.exp(-X))/(np.exp(X)+np.exp(-X))
Y=[round(i,2) for i in Y]
print(Y)
```

```python
#Relu Activation function
#Rectified Linear Unit 
Y=np.where(X>0,X,0)
#crate a dictionary 
dict={'X':X,'Y':Y}
#convert dictionary into DataFrame 
df=pd.DataFrame(dict)
#line plot
sns.lineplot(data=df,x=X,y=Y)
plt.title("Relu Activation function")
plt.show()
'''
if(X>0):
    Y=X
else:
    Y=0

'''
```
