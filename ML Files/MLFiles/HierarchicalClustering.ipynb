```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel("University_CLustering.xlsx")
df.head()
```

```python
df.info()
```

```python
len(df["Univ"].value_counts())
```

```python
new_df=df.drop(["Univ","State"],axis=1)
new_df.head(2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
new_df.iloc[:]=scaler.fit_transform(new_df.iloc[:])
new_df.head()
```

```python
import scipy.cluster.hierarchy as sch
from scipy.cluster.hierarchy import linkage
```

```python
z=linkage(new_df)
plt.figure(figsize=(15,8))
plt.xlabel("Index")
plt.ylabel("Distance")
sch.dendrogram(z,
              leaf_rotation=0,
              leaf_font_size=10
             )
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=3)
clustering.fit(new_df)
```

```python
clustering.labels_
```

```python
df["Class"]=clustering.labels_
```

```python
df[df["Class"]==0]["Univ"]
```

```python
df[df["Class"]==1]["Univ"]
```

```python
df[df["Class"]==2]["Univ"]
```

```python
new_df.head(2)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
components = pca.fit_transform(new_df)
```

```python
print(len(new_df.columns))
print(components)
```

```python
print(pca.explained_variance_ratio_)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
pc0=components[:,0]
pc1=components[:,1]
pc2=components[:,2]
```

```python
plt.figure(figsize=(8,8))
ax=plt.axes(projection="3d")
ax.scatter(pc0,pc1,pc2,c=df["Class"],cmap="viridis")
for i,u in enumerate(df["Univ"]):
    ax.text(pc0[i],pc1[i],pc2[i],u)
plt.show()
```
