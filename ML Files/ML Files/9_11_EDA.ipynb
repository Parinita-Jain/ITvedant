```python
def identify_outliers(group):
    q1=group["price"].quantile(0.25)
    q3=group["price"].quantile(0.75)
    iqr=q3-q1
    uw=q3+1.5*iqr
    lw=q1-1.5*iqr
    outliers=group[(group["price"]<lw)|(group["price"]>uw)]
    return outliers[["make","price"]]
    
    
```

```python
# dont drop above indexes 
def replace_outlier_with_median(group):
    q1 = group["price"].quantile(0.25)
    q3 = group["price"].quantile(0.75)
    iqr = q3 - q1
    lw = q1 - 1.5 * iqr
    uw = q3 + 1.5 * iqr
    median_price = group["price"].median()
    
    # Replace outliers with median
    group["price"] = group["price"].apply(lambda x: median_price if x < lw or x > uw else x)
    return group

# Apply the function group-wise by 'make'
df = df.groupby("make").apply(replace_outlier_with_median).reset_index(drop=True)
```

```python
'''
Overfitting is there, so decrease number of features.
when training acc>testing acc -- overfitting -- high variance low bias
when training acc<testing acc -- underfitting -- high bias low variance
bias variance tradeoff -- low bias low variance
To deal with overfitting - regularization, crossvalidation, hyperparameter tuning
'''
```

```python
# Regularization is of 2 types :
# L1 (Lasso) : RSS(MSE)+lambda*(sum(abs(slopes))) # lambda : 200-1000
# L2 (Ridge) : RSS(MSE)+lambda*(sum((slopes)**2)) # lambda : 1-50
```

```python
from sklearn.linear_model import Lasso,Ridge
for i in range(200,1001):
    ls=Lasso(i)
    ls.fit(xtrain,ytrain)
    score_train=ls.score(xtrain,ytrain)
    score_test=ls.score(xtest,ytest)
    print(f"{i}  {score_train}  {score_test}")
```

```python
ls=Lasso(alpha=223)
ls.fit(xtrain,ytrain)
ypred=ls.predict(xtest)
f=pd.DataFrame({"Columns":xtrain.columns,
                "Ls Coefficient":ls.coef_,
                "Linear Regress Coeff": lr.coef_})
print(f)
```

```python
for i in range(1,51):
    lri=Ridge(i)
    lri.fit(xtrain,ytrain)
    score_train=lri.score(xtrain,ytrain)
    score_test=lri.score(xtest,ytest)
    print(f"{i}  {score_train}  {score_test}")
```

```python
lri=Ridge(alpha=33)
lri.fit(xtrain,ytrain)
ypred=lri.predict(xtest)
f["Ridge Coeffs"]=lri.coef_
print(f)
```
