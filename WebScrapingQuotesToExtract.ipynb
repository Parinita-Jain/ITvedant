```python
# Install required libraries in Colab
!pip install requests
!pip install beautifulsoup4
!pip install pandas

# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape Snapdeal products for a search term
def scrape_snapdeal(search_term, num_pages=1):
    base_url = "https://www.snapdeal.com/search?keyword=hp%20laptop&santizedKeyword=laptops&catId=150&categoryId=0&suggested=false&vertical=p&noOfResults=20&searchState=&clickSrc=go_header&lastKeyword=&prodCatId=&changeBackToAll=false&foundInAll=false&categoryIdSearched=&cityPageUrl=&categoryUrl=&url=&utmContent=&dealDetail=&sort=rlvncy"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36"
    }

    products = []

    for page in range(1, num_pages + 1):
        params = {
            "keyword": search_term,
            "santizedKeyword": search_term,
            "catId": "",
            "page": page
        }
        response = requests.get(base_url, headers=headers, params=params)
        soup = BeautifulSoup(response.text, "html.parser")

        # Find all product containers
        product_cards = soup.find_all("div", class_="product-tuple-listing")

        for card in product_cards:
            title = card.find("p", class_="product-title")
            price = card.find("span", class_="product-price")
            rating = card.find("div", class_="filled-stars")

            products.append({
                "Title": title.text.strip() if title else None,
                "Price": price.text.strip() if price else None,
                "Rating": rating["style"].strip() if rating else None
            })

    return pd.DataFrame(products)

# Example: Scrape first 2 pages of "laptops"
df = scrape_snapdeal("laptops", num_pages=2)
df.head()

# Save to CSV
df.to_csv("snapdeal_laptops.csv", index=False)
print("Scraping completed! Data saved to snapdeal_laptops.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
#wikiurl="https://www.snapdeal.com/products/men-sports-footwear"
wikiurl="https://quotes.toscrape.com"
response=requests.get(wikiurl)
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

quotes_data = []

# Find all quote containers
quote_divs = soup.find_all('div', class_='quote')

for quote_div in quote_divs:
    text = quote_div.find('span', class_='text').text.strip()
    author = quote_div.find('small', class_='author').text.strip()
    tags_elements = quote_div.find_all('a', class_='tag')
    tags = [tag.text.strip() for tag in tags_elements]

    quotes_data.append({
        'Quote': text,
        'Author': author,
        'Tags': ', '.join(tags)
    })

df_quotes = pd.DataFrame(quotes_data)
print(df_quotes.head())
```
