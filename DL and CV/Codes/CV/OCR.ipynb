```python
img1= cv2.imread('img1.png')

# By default OpenCV stores images in BGR format and since pytesseract assumes RGB format,
# we need to convert from BGR to RGB format/mode:

img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) # open cv reads the image from bgr to rgb
result = pytesseract.image_to_string(img1_rgb)
```

```python
# create a new folder in colab named invoices and put all the 3 invoices inside it.
import os
import pandas as pd

invoices = os.listdir("invoices")
invoices
```

```python
prices = []
dates = []
for invoice in invoices:
    invoice_img = cv2.imread("invoices/"+invoice)
    invoice_rgb = cv2.cvtColor(invoice_img, cv2.COLOR_BGR2RGB)
    df = pytesseract.image_to_data(invoice_rgb, output_type="data.frame")
    dates.append(df["text"][43])
    prices.append(df["text"][119])

invoice_df = pd.DataFrame({"date":dates,"price":prices})
```
