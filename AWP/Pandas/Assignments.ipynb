```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
d={"age_groups":["Children (0-18)","Adults (19-25)","Adults (26-34)","Adults (35-54)","Adults (55-64)","Seniors (65+)"],
"populations":[75307800,27799100,39817700,81478600,42061700,52784400]}
df=pd.DataFrame(d)
colors=["skyblue","orange","green","pink","yellow","purple"]
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,10))
ax1.pie(df["populations"],labels=df["age_groups"],autopct="%1.2f%%",colors=colors,explode=[0,0,0.1,0,0,0])
ax1.set_title("Pie Chart : Population Distribution by Age Groups")
ax2.pie(df["populations"],labels=df["age_groups"],colors=colors,wedgeprops={"width":0.5},autopct="%1.2f%%")#"%2.1f%%"
ax2.set_title("Donut Chart : Population Distribution by Age Groups")
plt.show()
```

```python
# 2 Linear Regression Model
data={
    "weight":[140,155,159,179,192,200,212],
    "height":[60,62,67,70,71,72,75]
}
df=pd.DataFrame(data)
sns.regplot(data=df,x="height",y="weight")
plt.title("Regression Plot : Height vs Width")
plt.xlabel("Height")
plt.ylabel("Width")
plt.show()
```
