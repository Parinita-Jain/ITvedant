```python
# Feature engineering --  
df["Total"]=df["Python"]+df["Machine Learning"]
df.head(5)
```

```python
c=["Mumbai","Mumbai","Mumbai","Pune","Delhi","Delhi","Delhi","Delhi","Pune","Banglore"]
df["City"]=c
df.head()
```

```python
'''
'''
def f(x):
    return x/2
df["Total2"]=df["Total"].apply(f)
df["Total1"]=df["Total"].apply(lambda x:x/2)
df.head()
```

```python
def grade(marks):
    if marks>=75 and marks<100:
        return "A"
    elif marks>=60 and marks<75:
        return "B"
    elif marks>=35 and marks<60:
        return "C"
    else:
        return "F"
df["Grade"]=df["Total1"].apply(grade)   
df.head()
```
