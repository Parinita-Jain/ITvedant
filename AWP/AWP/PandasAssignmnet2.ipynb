```python
import pandas as pd #
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Netflix.csv")
df.head(2)
```

```python
# $20,00,000
print(df["Boxoffice"])
print(df["Boxoffice"].dtype)
```

```python
df["Boxoffice"]=df["Boxoffice"].str.replace("[$,]+","",regex=True).astype("float")
```

```python
box_office_hits=df[(df["Series or Movie"]=="Movie")&(df["Boxoffice"]>=2000000)]
print("Number of box_office hits : ",len(box_office_hits))
```

```python
#--
critically_acclaimed=df[(df["Series or Movie"]=="Movie")&(df["IMDb Score"]>=8)]
print("Number of critically acclaimed Movies : ",len(critically_acclaimed))
```
